```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np
import matplotlib.pyplot as mplt
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
df = pd.read_json(r'C:\Users\Usman\Desktop\Project\bjp.json', lines=True)
```

```python
dict = {
    'text':[text for text in df['full_text']],
    'compound':[sia.polarity_scores(text)['compound'] for text in df['full_text']],
    'negative':[sia.polarity_scores(text)['neg'] for text in df['full_text']],
    'positive':[sia.polarity_scores(text)['pos'] for text in df['full_text']],
    'neutral' :[sia.polarity_scores(text)['neu'] for text in df['full_text']]
}
```

```python
area = np.pi * 3
ap = np.arange(500)
mplt.scatter(dict['compound'][:3000],np.arange(3000), s=area, alpha=1)
mplt.title('Tweet Intensity for BJP')
mplt.xlabel('Sentiment Intensity')
mplt.ylabel('Frequency of Tweets')
mplt.show()
```
